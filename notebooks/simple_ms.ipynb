```python
def canonical_channel_name(x: str) -> str:
    x = str(x).strip().lower()
    for ch in ["&", "-", " ", "/", "\\", ".", ",", ":"]:
        x = x.replace(ch, "_")
    while "__" in x:
        x = x.replace("__", "_")
    return x.strip("_")
```
